# The variables

**Companies:** \\
Refers to the name or a list of companies being analyzed or tracked.

**URL:** \\
The web address (link) to the company's official website or other relevant web pages providing information.

**Description:** \\
A brief overview of what the company does, its core services/products, and its mission.

**Business Model:** \\
Describes how the company generates revenue. It outlines the company's approach to creating, delivering, and capturing value (e.g., subscription-based, freemium, B2B, B2C).

**Industry:** \\
The broader economic sector to which the company belongs (e.g., Technology, Healthcare, Finance).

**Sub-Industry:** \\
A more specific classification within the main industry (e.g., within Technology: FinTech, HealthTech).

**Expert Collections:** \\
A curated list of companies that experts have grouped based on themes, trends, or sectors of interest, providing further insights into market positioning.

**Markets:** \\
Refers to the geographical or vertical markets in which the company operates or serves (e.g., North America, Asia-Pacific, Consumer Goods).

**Mosaic (Overall):** \\
A composite score or rating often based on different metrics that evaluate a company's overall performance or health. It could involve financial, operational, and innovation factors.

**Commercial Maturity:** \\
The stage of the company's development in terms of market penetration and revenue generation (e.g., startup, growth stage, established).

**IPO Probability:** \\
The likelihood of the company going public via an Initial Public Offering (IPO) based on factors such as market positioning, financial health, and investor interest.

**M&A Probability:** \\
The likelihood that the company will be involved in a merger or acquisition in the future, considering its market position and industry trends.

**Country:** \\
The country where the company is headquartered or primarily operates.

**Total Funding (M):** \\
The total amount of money the company has raised through funding rounds, expressed in millions (M), typically from venture capital, private equity, or other investors.

**All Investors:** \\
A list of all the entities or individuals that have invested in the company across all funding rounds.

**Latest Funding Amount (M):** \\
The most recent amount of money raised by the company in its latest funding round, in millions.

**Latest Funding Round:** \\
The most recent stage of fundraising (e.g., Series A, Series B, Seed) the company has completed.

**Latest Funding Date:** \\
The date when the company closed its latest funding round.

**Total Headcount:** \\
The total number of employees working at the company.

**12 Month Headcount Change %:** \\
The percentage change in the company's headcount over the last 12 months, indicating growth or reduction in workforce.

**Company Status:** \\
Indicates the current operational state of the company (e.g., active, defunct, acquired, public).

**Founded Year:** \\
The year the company was established.

**Your Collections:** \\
Customized groups of companies or data points that the user has organized for their analysis or tracking purposes.

**Team Tag:** \\
An internal label or category assigned to a company's team members, often used for classification within an organization (e.g., leadership, development team).

Modification of the dataset: we want to remove the '%', as they are not useful for further computations.

```
def myfunc(x):
  for element in x:
    new = float(element.replace('%', ''))
    x= x.replace(element,new)
  return x

# Clean market names by removing parentheses and standardizing technology categories
def clean_market_name(market):
    # Remove everything in parentheses and strip whitespace
    cleaned = market.split('(')[0].strip()
    # If it contains 'Tech' or 'Technology', just return 'Technology'
    if 'Tech' in cleaned or 'Technology' in cleaned:
        return 'Technology'
    return cleaned
```

```
data['12 Month Headcount Change %']= data[['12 Month Headcount Change %']].apply(myfunc)
data['IPO Probability']=data[['IPO Probability']].apply(myfunc)
data['M&A Probability']=data[['M&A Probability']].apply(myfunc)
data.loc[:, 'Markets'] = data['Markets'].apply(clean_market_name)
```

```
import re
from collections import Counter
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize

def summarize_text(text, num_sentences=3):
    # Download necessary NLTK data (run once)
    import nltk
    nltk.download('punkt')
    nltk.download('stopwords')

    # Tokenize the text into sentences
    sentences = sent_tokenize(text)

    # Tokenize words and remove stopwords
    stop_words = set(stopwords.words('english'))
    words = word_tokenize(text.lower())
    words = [word for word in words if word.isalnum() and word not in stop_words]

    # Calculate word frequencies
    word_freq = Counter(words)

    # Calculate sentence scores based on word frequencies
    sentence_scores = {}
    for i, sentence in enumerate(sentences):
        for word in word_tokenize(sentence.lower()):
            if word in word_freq:
                if i in sentence_scores:
                    sentence_scores[i] += word_freq[word]
                else:
                    sentence_scores[i] = word_freq[word]

    # Get the top n sentences with the highest scores
    top_sentences = sorted(sentence_scores, key=sentence_scores.get, reverse=True)[:num_sentences]

    # Combine the top sentences in their original order
    summary = ' '.join([sentences[i] for i in sorted(top_sentences)])

    return summary
```

```
def create_chart(mylist,title):
    # Filter data for the markets in mylist
    filtered_data = data[data['Markets'].isin(mylist)]

    # Get unique markets and their counts
    market_counts = filtered_data['Markets'].value_counts()
    mywords = market_counts.index.tolist()
    numbers = market_counts.values.tolist()

    # Create figure and axis
    fig, ax = plt.subplots(figsize=(14, 6))

    # Create horizontal bars
    y_pos = np.arange(len(mywords))
    bars = ax.barh(y_pos, numbers, align='center')

    # Color each bar differently
    colors = plt.cm.viridis(np.linspace(0, 1, len(numbers)))
    for bar, color in zip(bars, colors):
        bar.set_color(color)

    # Summarize market names
    words = [summarize_text(word, num_sentences=1) for word in mywords]

    # Customize the plot
    ax.set_yticks(y_pos)
    ax.set_yticklabels(words)
    ax.invert_yaxis()  # Words read top-to-bottom
    ax.set_xlabel('Number of Occurrences')
    ax.set_title(title)

    # Set x-axis limits to include all data points
    ax.set_xlim(0, max(numbers) * 1.1)

    # Add value labels on the bars
    for i, v in enumerate(numbers):
        ax.text(v, i, str(v), va='center', ha='left', fontweight='bold', fontsize=10, color='black')

    # Remove y-axis line
    ax.spines['left'].set_visible(False)

    # Create legend
    legend_elements = [plt.Rectangle((0,0),1,1, fc=color, label=f'{word} ({num})')
                       for color, word, num in zip(colors, mywords, numbers)]
    ax.legend(handles=legend_elements, loc='center left', bbox_to_anchor=(1, 0.5),
              title="Legend", title_fontsize=12)

    # Adjust layout and display the plot
    plt.tight_layout()
    plt.show()
```
